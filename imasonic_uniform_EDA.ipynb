```python
mat = scipy.io.loadmat(os.path.join(data_path, 'homogenous/rf_imasonic_uniform.mat'))
rf = mat['rf']
line = rf[:6000, 0].T
print(line.shape)
f, t, Sxx = signal.spectrogram(line, fs=FS, nperseg=256)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim([0, 2e7])
# plt.xlim([0, 0.00006])
plt.title('AC 0.7')
plt.show()  
```
