```python
import matplotlib
font = {'family' : 'normal',
        'size'   : 15}

matplotlib.rc('font', **font)

fig, ax = plt.subplots()
ax.set_ylabel("AC ($dB / (MHz \\times cm)$)")
ax.set_xlabel("$k$")
fig.set_size_inches(10, 9)
ax.boxplot([data_02cm, data_05cm, data_1cm], labels=['0.2 cm', '0.5 cm', '1 cm'], showmeans=True, 
           meanprops = dict(marker='D', 
                markeredgecolor='black',
                markerfacecolor='firebrick'),
           meanline=False)
ax.axhline(y=0.7, color='b', linestyle='--', linewidth=0.5)
plt.savefig('ius19_boxplots.pdf', dpi=300)
# ax.set_yticks(list(ax.get_yticks()) + np.array([0.7]))
```
